## KNN 구현

```python
X, y  = mglearn.datasets.make_forge()
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
print(X_test.shape, X_train.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=3)
clf.fit(X_train, y_train)
```

```python
# 학습(훈련) 결과 테스트
clf.predict(X_test)

# 테스트 정확도 측정
clf.score(X_test, y_test)
```

### 실제 데이터 적용


```python
from sklearn.datasets import load_breast_cancer #유방암 데이터

cancer=load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=66, 
                                                    stratify=cancer.target) # 암걸린사람의 비율이 5:5가 될수가 없으니 테스트 데이터도 맞추어서 나누어준다 stratify

print(X_train.shape, X_test.shape)
```

```python
clf = KNeighborsClassifier(n_neighbors=6)
clf.fit(X_train, y_train)

#Train set의 정확도
print(clf.score(X_train, y_train))

#test set의 정확도
print(clf.score(X_test, y_test))
```

```python
train_acc=[]
test_acc=[]
max_n_neigh=10
for n_neigh in range(1,max_n_neigh):
    clf = KNeighborsClassifier(n_neighbors=n_neigh)
    clf.fit(X_train, y_train)

    #Train set의 정확도
    train_acc.append(clf.score(X_train, y_train))

    #test set의 정확도
    test_acc.append(clf.score(X_test, y_test))
plt.plot(range(1,max_n_neigh),train_acc, label="훈련 정확도")
plt.plot(range(1,max_n_neigh),test_acc, label="테스트 정확도")
plt.ylabel("정확도")
plt.xlabel("이웃의 갯수")
plt.legend()
```

## 영화 분류 프로그램 만들기

```python
import operator
def createDataSet():
    group = np.array([[1.0,2.0],
                            [1.0,4.0],
                            [4.0,2.0],
                            [4.0,2.0]])
    label = ["Action","Action","Romance","Romance"]
    return group,label

def calcDistance(x_new,data):
    dataSize=data.shape[0]                 #데이터 갯수
    diffMat = np.tile(x_new,(dataSize,1))  - data            # x_new 데이터를 4행 1열로
    
    # L2 norm
    sqdiffMat = (diffMat**2)
    sqDistance=sqdiffMat.sum(axis=1)  #행 기준 연산
    distance = sqDistance**0.5
    
    sortedistance = distance.argsort()
    return sortedistance

def findMajorityClass(x_new,target,k,sortDistance):
    classCount = {}
    
    for i in range(k):
        vortelabel= target[sortDistance[i]]
        classCount[vortelabel] = classCount.get(vortelabel,0)+1
    
    return sorted(classCount.items(),key=operator.itemgetter(1), reverse=True)
        
def classify(x_new,data,target,k):
    sortDistance=calcDistance(x_new,data) # retrun 값 0, 1, 2, 3
    sortedDistanceCount=findMajorityClass(x_new,target,k,sortDistance)
    
    return sortedDistanceCount
```

```python
X,y = createDataSet()
result=classify([2.0,3.0],X,y,3)
print("결과는 {} 영화다".format(result[0][0]))
```
